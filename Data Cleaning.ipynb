```python
#import modules and libraries
import pandas as pd
from glob import glob
```

## Use glob() to list all files that match a pattern and sort results

```python
stock_files = sorted(glob('data/UNSW-NB15_*.csv'))
stock_files
```

### Build data frame from each csv file and store all data frames in one list

```python
dataframes = []
for file in stock_files:
    dataframes.append(pd.read_csv(file))
```

### Getting column names from the featues file the assign it to each data frame in the list we created above

```python
features = pd.read_csv("NUSW-NB15_features.csv", usecols=[1])
features_str = list(features.values.flatten())
for df in dataframes:
    df.columns = features_str
```

### Combining all dataframes into a big one

```python
all_data = pd.concat(dataframes)
```

```python
all_data['attack_cat'] = df['attack_cat'].fillna('Normal')
all_data.attack_cat.unique()
```

### Get DoS attacks and Normal flows only from the all data

```python
normal = all_data.loc[all_data['attack_cat'] == 'Normal'].sample(n=1000000)
dos = all_data.loc[all_data['attack_cat'] == 'DoS']
result = pd.concat([normal, dos])
result
```

```python
result.to_excel("result.xlsx")
```
